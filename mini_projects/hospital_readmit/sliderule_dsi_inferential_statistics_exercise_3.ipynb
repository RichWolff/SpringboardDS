# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

First, I'll test the difference of means between >1000 discharge hospitals less the mean <100 discharge hospitals.

# Is there a difference between >1000 discharge hospitals and those with <100 discharges?
#### Null hypothesis: There is a difference in excessRatio between hospitals with >1000 discharges and <100 discharges
#### T statistic: Difference of means, under 100 discharges mean less than over 1000 discharges mean


```python
size = 10000
replicates = np.empty(size)
over1k = discharges>1000
under100 = discharges<100
orgDiff = excessRatio[under100].mean()-excessRatio[over1k].mean()
excessDataConcat = np.concatenate([excessRatio[over1k],excessRatio[under100]])

for i in range(size):
    permData = np.random.permutation(excessDataConcat)
    permLess100 = permData[np.sum(over1k):]
    permMore1000  = permData[:np.sum(over1k)]
    replicates[i] = np.mean(permLess100) - np.mean(permMore1000)
pval = np.sum(replicates>=orgDiff)/len(replicates)

plt.hist(replicates,bins=30,normed=True,color='#0011CC')
plt.vlines(orgDiff,0,85,color='red')

marginError = stats.norm.ppf(.995)*replicates.std()
low = replicates.mean() - marginError
high = replicates.mean() + marginError
plt.vlines(low,0,40,color='#CC1100',alpha=.5)
plt.vlines(high,0,40,color='#CC1100',alpha=.5)
plt.show()

print("Margin Of Error at 99% confidence interval: {0:0.5f}".format(marginError))
print("Conf Bands: ({0:.5f}, {1:.5f})".format(low,high))
print("Observed Difference: {0:0.5f}".format(orgDiff))
print('P Value: {0:0.5f}'.format(pval))
print('Percent of sample not included in test: {0:0.2f}'.format((len(excessRatio)-len(excessDataConcat))/len(excessRatio)))
```

## There does indeed seem to be a negative correlation between readmission and discharges

<b> Next, i'd like to test for significance using the pearson correlation stat. </b>

<b>Null Hypothesis:</b> Discharges and excessive readmissions are completely uncorrelated<br>
<b>Test Statistic:</b> Pearson correlation

```python
## To test this hypothesis, we'll test for independence from each other
## we'll leave excess ratio the same and permute discharges
r_obs = np.corrcoef(excessRatio,discharges)[0,1]
perms = 10000
perm_correlations = np.empty(perms)
for i in range(perms):
    temp_discharges_permuted = np.random.permutation(discharges)
    perm_correlations[i] = np.corrcoef(excessRatio,temp_discharges_permuted)[0,1]

plt.hist(perm_correlations,bins=30)
plt.vlines(r_obs,0,1000,color='r')
plt.title('Simulated ')
plt.xlabel('Pearson Coefficients')
plt.ylabel('Occurrences')

marginError = stats.norm.ppf(.995)*perm_correlations.std()
low = perm_correlations.mean() - marginError
high = perm_correlations.mean() + marginError
pval = np.sum(perm_correlations<r_obs)/len(perm_correlations)
plt.vlines(low,0,500,color='#CC1100',alpha=.5)
plt.vlines(high,0,500,color='#CC1100',alpha=.5)

plt.show()

print("Margin Of Error at 99% confidence interval: {0:0.5f}".format(marginError))
print("Conf Bands: ({0:.5f}, {1:.5f})".format(low,high))
print("Observed Difference: {0:0.5f}".format(orgDiff))
print('P Value: {0:0.5f}'.format(pval))
print('Percent of sample not included in test: {0:0.2f}'.format((len(excessRatio)-len(excessDataConcat))/len(excessRatio)))

#With a p value of close to zero, 
#there is no doubt that these two variables are correlaed, 
#albeit a very low correlation
```

## For my last test I want to see if there is any difference in the excess readmission rate between the lower 50% of discharges and the upper 50%. 

<B>Null Hypothesis:</B> There is no difference in the excess readmition rate between the lower and upper 50% of hospital discharges.

<B>T Stat:</B> excess readmit rate mean of the lower 50% discharges less the excess readmit rate mean of the upper 50% discharges

```python
size = 10000
replicates = np.empty(size)
lower50 = excessRatio[discharges<np.median(discharges)]
upper50 =excessRatio[discharges>=np.median(discharges)]
orgDiff = lower50.mean()-upper50.mean()
excessDataConcat = np.concatenate([lower50,upper50])

for i in range(size):
    permData = np.random.permutation(excessDataConcat)
    permLess50 = permData[:len(lower50)]
    permMore50  = permData[len(lower50):]
    replicates[i] = np.mean(permLess50) - np.mean(permMore50)
pval = np.sum(replicates>=orgDiff)/len(replicates)

plt.hist(replicates,bins=30,normed=True,color='#0011CC')
plt.vlines(orgDiff,0,225,color='red')

marginError = stats.norm.ppf(.995)*replicates.std()
low = replicates.mean() - marginError
high = replicates.mean() + marginError
plt.vlines(low,0,100,color='#CC1100',alpha=.5)
plt.vlines(high,0,100,color='#CC1100',alpha=.5)
plt.show()

print("Margin Of Error at 99% confidence interval: {0:0.5f}".format(marginError))
print("Conf Bands: ({0:.5f}, {1:.5f})".format(low,high))
print("Observed Difference: {0:0.5f}".format(orgDiff))
print('P Value: {0:0.5f}'.format(pval))
```
